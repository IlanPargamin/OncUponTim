```python
from transformers import (AutoTokenizer, 
                          AutoModelForSequenceClassification, 
                          TextClassificationPipeline)
```

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification, TextClassificationPipeline

tokenizer = AutoTokenizer.from_pretrained('roberta-base')
model = TFAutoModelForSequenceClassification.from_pretrained("ilan541/OncUponTim")

pipe = TextClassificationPipeline(model=model, tokenizer=tokenizer, return_all_scores=True)
```

# Explainer with SHAP

```python
def score_and_visualize(text):
  prediction = pipe([text])
  print(f"score bad text: {prediction[0][0]['score']:4f}\nscore good text: {prediction[0][1]['score']:4f}")

  explainer = shap.Explainer(pipe)
  shap_values = explainer([text])

  shap.plots.text(shap_values)
```

```python
score_and_visualize('''My mother drove me to the airport with the windows rolled down. It was seventy-five degrees in
Phoenix, the sky a perfect, cloudless blue. I was wearing my favorite shirt — sleeveless, white eyelet
lace; I was wearing it as a farewell gesture. My carry-on item was a parka.!''')
```

```python
#@markdown ---
your_text = 'My mother drove me to the airport with the windows rolled down. ' #@param {type:"string"}

score_and_visualize(your_text)
#@markdown ---
```
